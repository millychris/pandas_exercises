# US - Baby Names

### Introduction:

We are going to use a subset of [US Baby Names](https://www.kaggle.com/kaggle/us-baby-names) from Kaggle.  
In the file it will be names from 2004 until 2014


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv). 

### Step 3. Assign it to a variable called baby_names.

```python
url='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv'
babyname= pd.read_csv(url, sep=',')
```

### Step 4. See the first 10 entries

```python
babyname.head(10)
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
del babyname['Unnamed: 0']


del babyname['Id']

babyname.head()
```

### Step 6. What year has the highest number of baby names in the dataset?

```python
babyname['Name'].value_counts()
```

```python
babyname.groupby("Year").count()
```

```python
babyname.groupby("Year")["Name"].count().sort_values(ascending=False)[:1]
```

### Step 7. Is there more male or female names in the dataset?

```python
babyname['Gender'].value_counts()
```

### Step 8. Group the dataset by name and assign to names

```python
names = babyname.groupby("Name")['Count'].sum()
names[:10]
```

### Step 9. How many different names exist in the dataset?

```python
names.count()
```

### Step 10. What is the name with most occurrences?

```python
names.sort_values(ascending=False)
```

### Step 11. How many different names have the least occurrences?

```python
len(names[names == names.min()])
```

```python
names[names == names.min()]
```

```python
names[names == names.min()].count()
```

### Step 12. What is the median name occurrence?

```python
names[names == names.median()]
```

### Step 13. What is the standard deviation of names?

```python
names.std()
```

### Step 14. Get a summary with the mean, min, max, std and quartiles.

```python
names.describe()
```
