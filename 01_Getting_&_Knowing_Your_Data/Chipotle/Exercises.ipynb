### Step 10. For the most-ordered item, how many items were ordered?

```python
chipo['item_name'].dtype
ordered=chipo.groupby('item_name')
ordered_sum= ordered['quantity'].sum()
ordered_sum.sort_values(ascending=False)[:1]
```

### Step 11. What was the most ordered item in the choice_description column?

```python
chipo_exploded = chipo.copy()
chipo_exploded['choice_description'] = chipo['choice_description'].str.replace('[', '').str.replace(']', '').str.split(',')
chipo_exploded = chipo_exploded.explode('choice_description')
chipo_exploded['choice_description'] = chipo_exploded['choice_description'].str.strip()
chipo_exploded.groupby('choice_description')['quantity'].sum().sort_values(ascending=False)
```
