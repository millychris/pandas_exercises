### Step 4. How many products cost more than $10.00?

```python
# 假設附餐會影響到價格
chipo[chipo['item_name'] == 'Chicken Bowl'][['item_name', 'choice_description', 'item_price', 'quantity']].head(10)
```

```python
# 去除quantity的影響
chipo['price']=chipo['item_price'].str.replace('$', '').astype(float)
chipo['unit_price'] = chipo['price'] / chipo['quantity'].astype(float)
chipo['unit_price']
```

```python
# 建立multiIndex，做兩次groupby
product= chipo.groupby(['item_name', 'choice_description'])['unit_price']
product.size()
```

```python
# 就算把choice_description字串爆開，大概也算不出各配菜單價:(
# 「product」在這題的定義就是 item_name + choice_description 的唯一組合!不是光看 item_name。
```

```python
series_product=product.first()
```

```python
series_product.between(0,10)
```

```python
series_product > 10
```

```python
(series_product > 10).sum()
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive_item=chipo.sort_values(['unit_price'],ascending= False)[:1]
most_expensive_item
```

```python
chipo.sort_values(['unit_price'], ascending=False)[:1]['quantity']
```
